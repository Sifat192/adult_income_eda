# Adult Income Dataset - Exploratory Data Analysis

## Objective
The goal of this project is to perform Exploratory Data Analysis (EDA) on the Adult Income dataset to understand the factors that influence whether an individual's income exceeds $50K per year.

## Dataset Description
The dataset contains demographic and employment-related information such as age, education, occupation, hours worked per week, and marital status. The target variable indicates whether a person's income is greater than $50K annually.

## Analysis Goals
- Understand the distribution of demographic features
- Explore relationships between education, occupation, and income
- Identify patterns that influence income levels
- Prepare the dataset for potential machine learning models

## 4. Exploratory Data Analysis (EDA) & Data visualizations

```python
import numpy as np
df = df.replace('?',np.nan)
```

```python
df.isin(['?']).sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(how = 'any', inplace = True)
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.describe(include = 'all')
```

```python
df = df.drop(['educational-num','capital-gain','capital-loss'],axis = 1)
```

```python
import matplotlib.pyplot as plt 
```

```python
df['age'].hist()
plt.xlabel = 'Age'
plt.ylabel = 'Number of People'
plt.title = 'Adult Age Distribution'
```

```python
df['age'].between(17,48).sum()
```

```python
df.workclass
```

```python
plt.figure(figsize= (9,8))
df['workclass'].hist()
```

```python
df[df['education']=='Bachelors'].count()
```

```python
df[df['education']=='Masters'].count()
```

```python
df['income']=df['income'].replace('<=50K',0)
df['income']=df['income'].replace('>50K',1)
```

## 5.Conclusions

In this project, we performed Exploratory Data Analysis on the Adult Income dataset to understand the factors that influence whether an individual's income exceeds $50K per year.

Through the analysis, we observed that demographic and employment-related features such as age, education level, occupation, marital status, and hours worked per week show noticeable patterns in relation to income levels. Certain education levels and occupations appear more frequently among individuals earning higher incomes, while working hours also show a relationship with income distribution.

The exploratory analysis helped in identifying important variables and understanding the structure of the dataset. These insights are useful for preparing the data for machine learning models that could predict income categories.
